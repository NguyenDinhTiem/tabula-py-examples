# Trích xuất bảng

## Bảng có cấu trúc rõ ràng

Các bảng này có đường viền rõ ràng xung quanh cả bảng và các ô bên trong, giúp dễ dàng nhận diện và trích xuất.
Chúng thường được tạo ra từ các ứng dụng như Microsoft Word hoặc Excel trước khi chuyển đổi sang định dạng PDF.

```python
import tabula
pdf_path = "https://github.com/NguyenDinhTiem/tabula-py-examples/raw/main/data/tables.pdf"

dfs = tabula.read_pdf(pdf_path, pages="1", lattice=True)
print(len(dfs))
dfs[0]
```

## Bảng không có đường viền

Các bảng này không có đường viền(hoặc có đường viền không đầy đủ) xác định rõ ràng giữa các hàng và cột, chỉ dựa vào khoảng cách để phân chia các ô.
Việc trích xuất từ các bảng này có thể khó khăn hơn do thiếu các dấu hiệu trực quan rõ ràng.


```python
import tabula
pdf_path = "https://github.com/NguyenDinhTiem/tabula-py-examples/raw/main/data/tables.pdf"

dfs = tabula.read_pdf(pdf_path, pages="2", stream=True)
print(len(dfs))
dfs[0]
```

## Bảng có cấu trúc phức tạp

Bao gồm các bảng có hàng hoặc cột được gộp, các bảng con, hoặc các bảng chứa nhiều cấp độ của thông tin.
Các bảng này thường xuất hiện trong các tài liệu kỹ thuật, tài chính hoặc khoa học, và việc trích xuất dữ liệu từ chúng đòi hỏi xử lý đặc biệt.

```python
import tabula
pdf_path = "https://github.com/NguyenDinhTiem/tabula-py-examples/raw/main/data/tables.pdf"

dfs = tabula.read_pdf(pdf_path, pages="3", stream=True)
print(len(dfs))
dfs[0]
```

```python
df = dfs[0]
# In DataFrame gốc
print("DataFrame gốc:")
print(df)

# Điền các giá trị NaN bằng các giá trị từ hàng trước đó
df['Môn học'].fillna(method='ffill', inplace=True)
df['Ngày thi'].fillna(method='ffill', inplace=True)
df['Giờ bắt đầu'].fillna(method='ffill', inplace=True)
df['Phòng thi'].fillna(method='ffill', inplace=True)

print("\nDataFrame sau khi fillna:")
print(df)

# Loại bỏ những hàng sai 0, 3, 6
df_cleaned = df.drop([0, 3, 6])

# In DataFrame sau khi làm sạch
print("\nDataFrame sau khi làm sạch:")
print(df_cleaned)
```
